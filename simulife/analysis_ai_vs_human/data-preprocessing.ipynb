```python
def parse_options(data_section):
    return data_section["story"] + data_section["question"] if "question" in data_section else "", data_section["option_1"] if "option_1" in data_section else "", data_section["option_2"] if "option_2" in data_section else "", data_section["option_3"] if "option_3" in data_section else "", data_section["choice"] if "choice" in data_section else ""

    
def parse_story(data_section):
    # print((data_section["story"]))
    return data_section["story"]  if "story" in data_section else ""
    
def parse_converation(data_section):
    # print((data_section["messages"]))
    return data_section["messages"] if "messages" in data_section else ""

def get_sage_data(sage_data, data, index):
    _id = data[index]["_id"]
    print(_id)
    for document in sage_data:
        if document["scriptplay_id"] == _id:
            return document
    return ""
```

```python
def parse(collecton):
    users = []
    story_data = []
    decision_making_data = []
    conversation_data = []
    group_data = []
    sage_data = []
    
    for name, document_index in zip(no_sage_names, no_sages):
        document = data[document_index]
        # print(name, "   ", document_index)
        decision_of_user = []
        story_of_user = []
        conversation_of_user = []
        group_of_user = []
        
        for i in range(len(document["playdata"])):
            # print(document_index, " -- ", i)
            if document["playdata"][i]["event_type"] == "generate-options":
                question, option1, option2, option3, user_choice = parse_options(document["playdata"][i])
                decision_of_user.append({"question":question, "option1":option1, "option3":option3, "user_choice":user_choice})
                    

            story_of_user.append(parse_story(document["playdata"][i]))

            if document["playdata"][i]["event_type"] == "generate-character":
                 conversation_of_user.append(parse_converation(document["playdata"][i]))
            
            if document["playdata"][i]["event_type"] == "generate-groupchat":
                 group_of_user.append(parse_converation(document["playdata"][i]))
            
                    
                    
        users.append(name)
        story_data.append(story_of_user)
        decision_making_data.append(decision_of_user)
        conversation_data.append(conversation_of_user)
        group_data.append(group_of_user)
        sage_data.append("")
    df = pd.DataFrame({"Name":users, "Sage":"without", "Stroy_data":story_data, "Decision_making_data":decision_making_data, "Conversation_data":conversation_data, "Group_data":group_data, "Sage_data":sage_data})
    return df


def parse2(collecton):
    users = []
    story_data = []
    decision_making_data = []
    conversation_data = []
    group_data = []
    sage_data = []
    # sage_data = []
    
    for name, document_index in zip(no_sage_names, with_sages):
        document = data[document_index]
        # print(name, "   ", document_index)
        decision_of_user = []
        story_of_user = []
        conversation_of_user = []
        group_of_user = []
        # sage_of_user = []
        
        for i in range(len(document["playdata"])):
            # print(document_index, " -- ", i)
            if document["playdata"][i]["event_type"] == "generate-options":
                question, option1, option2, option3, user_choice = parse_options(document["playdata"][i])
                decision_of_user.append({"question":question, "option1":option1, "option3":option3, "user_choice":user_choice})
                    

            story_of_user.append(parse_story(document["playdata"][i]))

            if document["playdata"][i]["event_type"] == "generate-character":
                 conversation_of_user.append(parse_converation(document["playdata"][i]))
            
            if document["playdata"][i]["event_type"] == "generate-groupchat":
                 group_of_user.append(parse_converation(document["playdata"][i]))
                    
            # sage_of_user.append(get_sage_data(sage_data, data, document_index))
                    
           
        users.append(name)
        story_data.append(story_of_user)
        decision_making_data.append(decision_of_user)
        conversation_data.append(conversation_of_user)
        group_data.append(group_of_user)
        sage_data.append(get_sage_data(document_index))
        
    df = pd.DataFrame({"Name":users, "Sage":"with", "Stroy_data":story_data, "Decision_making_data":decision_making_data, "Conversation_data":conversation_data, "Group_data":group_data, "Sage_data":sage_data})
    return df
```

```python
def get_sage_ai(info):
    if type(info) == dict:
        return [msg for msg in info["messages"] if msg["role"] =="assistant"]
    else:
        return ""

    
def get_sage_human(info):
    if type(info) == dict:
        return [msg for msg in info["messages"] if msg["role"] != "assistant"]
    else:
        return ""
    
def get_sage_type(info):
    if type(info) == dict:
        return info["agent"]
    else:
        return ""
    
df["Sage_type"] = df["Sage_data"].apply(get_sage_type)
df["Sage_data(AI)"] = df["Sage_data"].apply(get_sage_ai)
df["Sage_data(Human)"] = df["Sage_data"].apply(get_sage_human)
```

```python
def get_conversation_ai(info):
    res = []
    for messages in info:
        res.append([msg for msg in messages if msg["role"] =="assistant"])
    return res
        

    
def get_conversation_human(info):
    res = []
    for messages in info:
        res.append([msg for msg in messages if msg["role"] != "assistant" and msg["role"] != "system"])
    return res

df["Conversation_data(AI)"] = df["Conversation_data"].apply(get_conversation_ai)
df["Conversation_data(Human)"] = df["Conversation_data"].apply(get_conversation_human)
```

```python
users_roles = ["Geralt", "Harry", "Lin"]

def get_group_ai(info):
    res = []
    for messages in info:
        res.append([msg for msg in messages if msg["speaker"] not in users_roles])
    return res
        

    
def get_grou_human(info):
    res = []
    for messages in info:
        res.append([msg for msg in messages if msg["speaker"] in users_roles])
    return res

df["Group_Conversation_data(AI)"] = df["Group_data"].apply(get_group_ai)
df["Group_Conversation_data(Human)"] = df["Group_data"].apply(get_grou_human)
```
